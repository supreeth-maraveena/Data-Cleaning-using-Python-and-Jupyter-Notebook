```python
#Step 1: Loading the Dataset

import pandas as pd
import numpy as np

df = pd.read_csv("marketing_campaign_data_messy.csv")
print(f"Loaded Dataset : {df.shape[0]} Rows and {df.shape[1]} Columns")
```

```python
df
```

```python
#Step 2: Header Cleaning

#1. Remove extra spaces at start and end of a Header using str.strip()
#2. Make all the Headers to lowercase using str.lower()
#3. Replace all the spaces in between using str.replace(' ', '_')

print(df.columns.tolist())

df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

print("FIX APPLIED")
print(df.columns.tolist())
```

```python
#Step 3: Type Conversion and Currency Cleaning
#We have to Remove any Currency Symbols from the data to perform any Numeric Operations

#1. Select all the rows which contain the $ symbol
#2. Replace that symbol with empty character using str.replace(r'[^\d.-]','')
#3. Then convert the data type from str back to numeric

#r'\$' → detects dollar symbol

#r'[^\d.-]' → detects all digits (\d), decimal point (.), and minus sign (-); 
#             the ^ (NOT) operator is used to ignore these valid characters.
#             Hence, it removes anything that is not a digit, ., or -.


dirty_spend = df['spend'].astype(str).str.contains(r'\$')
print(df.loc[dirty_spend,['campaign_id','spend']].head(3)) # before Currency Cleaning

df['spend'] = df['spend'].astype(str).str.replace(r'[^\d.-]', '', regex=True)
df['spend'] = pd.to_numeric(df['spend'])

print("FIX APPLIED")
print(df.loc[dirty_spend,['campaign_id','spend']].head(3)) # after Currency Cleaning
```

```python
#Step 4: Categorical Typos

#Here we check for human typos. Since typos refer to the same Category, We create a Dictionary to replace the values

print(df['channel'].unique()) # initially there were multiple names for same category

cleanup_map = {
    'Facebok' : 'Facebook',
    'Insta_gram' : 'Instagram',
    'Gogle' : 'Google Ads',
    'Tik_Tok' : 'TikTok',
    'E-mail' : 'Email',
    'N/A': np.nan
}

df['channel'] = df['channel'].replace(cleanup_map)

print("FIX APPLIED")
print(df['channel'].unique()) # all the typos has been removed
```

```python
print(df['active'].unique())

bool_map = {

            'Yes': True,
            'Y': True,
            '1': True,
            1: True,
            'No': False,
            '0': False,
            0: False
}

df['active'] = df['active'].map(bool_map).fillna(False).astype(bool)

print("FIX APPLIED")
print(df['active'].unique())
```

```python
#Step 6. Date Parsing
#First we check the datatype of all the date column and change it to 'datetime' datatype.

print(df['start_date'].dtype) # Initially they are of object type
print(df['end_date'].dtype)

df['start_date'] = pd.to_datetime(df['start_date'], dayfirst=True, errors='coerce')
df['end_date'] = pd.to_datetime(df['end_date'], dayfirst=True, errors='coerce')

print("FIX APPLIED")
print(df['start_date'].dtype) # now they became datetime64 datatype
print(df['end_date'].dtype)
```

```python
print(df['start_date'])
```

```python
#Step 7. Remove Duplicate Columns (This should be after step 2 and before step 3)

df = df.loc[:, ~df.columns.duplicated()]
df
```

```python
#Step 8: Logical Integrity (Clicks vs Impressions)
# They are the conditions b/w 2 columns or a column and a value that should be held True everywhere

impossible_mask = df['clicks'] > df['impressions']
print(df.loc[impossible_mask,['campaign_id','impressions','clicks']].head(3))
```

```python
#Step 8: Logical Integrity (Correct Dates)
# We check whether the start_date is before end_date or not. 
# if the above condition violates, we 1. remove that row
#                                     2. swap the start and end dates
#                                     3. add/subtract the duration to start date/from end date.

time_travel_mask = df['end_date'] < df['start_date']
print(df.loc[time_travel_mask,['campaign_id','start_date','end_date']].head(3))

df.loc[time_travel_mask, 'end_date'] = df.loc[time_travel_mask, 'start_date'] + pd.Timedelta(days=30)

print("FIX APPLIED")
print(df.loc[time_travel_mask,['campaign_id','start_date','end_date']].head(3))
```

```python
#Step 9: Handling Outliers (Winsorization)

Q1 = df['spend'].quantile(0.25)
Q3 = df['spend'].quantile(0.75)

IQR = Q3 - Q1 # IQR - Inter Quartile Range
upper_limit = Q3 + (3 * IQR)

outlier_mask = df['spend'] > upper_limit
print(df.loc[outlier_mask, ['campaign_id', 'spend']].head(3))

print("Fix Applied")

df.loc[outlier_mask, 'spend'] = upper_limit # I am equating all the outliers to the upper limit value
print(df.loc[outlier_mask, ['campaign_id', 'spend']].head(3))
```

```python
#Step 9: String Parsing (Feature Extraction)
#Here we will be extracting a certain part of the string from a column in a dataframe

print(df['campaign_name'].head(3))

df['season'] = df['campaign_name'].str.extract(r'Q\d_([^_]+)_')
#In the above command, r - regex
#   Q\d_([^_]+)_ is te structure of the data present in the columns
#   Q4_Summer_CMP-00001. Here \d = 4
#   the part in the ( ) is what we are extracting. ([^_]+) means anything except _

print("FIX APPLIED")

print(df[['campaign_name','season']].head(3))
```
